```python
import pandas as pd
import requests
import os
from py_edamam import Edamam
import getpass  
```

```python
url = "https://api.edamam.com/api/recipes/v2"

app_id = getpass.getpass()
app_key = getpass.getpass()

params = {
    'type':'public',
    'beta':'false',
    'app_id': app_id,
    'app_key': app_key,
    'cuisineType':'mediterranean', 
    'diet':'balanced',
    'random':'true'
    }

response = requests.request("GET", url, params=params)
# print(response.status_code)
# print(response.text)
```

```python
table = pd.DataFrame(response.json()['hits'])
table.head(3)
```

```python
def clean_list(row):
    return row.strip('[]').strip("'").split("', '")
```

```python
def clean_dict(row):
    x = row.strip('[]').split("}, {")
    x = [i.strip('{').split(", '") for i in x]
    xx = [i.split(": ") for i in x[0]]
    return {key.strip("'"):val.strip("'") for key, val in xx}
```

```python
recipes['dietLabels'] = recipes['dietLabels'].apply(clean_list)
```

```python
recipes['ingredients'] = recipes['ingredients'].apply(clean_dict)
```
