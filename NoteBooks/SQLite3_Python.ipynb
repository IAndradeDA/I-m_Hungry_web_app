## Create a DataBase

```python
# Create a database and open the database.
# If the database already exists just opens the database
conn = sqlite3.connect('../final_project.db')
c = conn.cursor()
```

## Create User Table

```python
# Create a users table if the table does not exists
c.execute('''CREATE TABLE IF NOT EXISTS recipes(
    recipe_id INTEGER PRIMARY KEY AUTOINCREMENT,
    label TEXT,
    images TEXT, 
    url TEXT,
    shareAs TEXT,
    yield REAL, 
    dietLabels TEXT, 
    healthLabels TEXT,
    cautions TEXT,
    ingredients TEXT,
    calories REAL, 
    cuisineType TEXT,
    mealType TEXT,
    dishType TEXT);
    ''')
# commit changes and close database connect
conn.commit()
```

```python
# Create another table
c.execute("""CREATE TABLE IF NOT EXISTS nutrition_table(
    food_id REAL PRIMARY KEY,
    food_name TEXT,
    level_1 TEXT,
    level_2 TEXT,
    level_3 TEXT,
    energy_kcal REAL,
    energy_kg REAL,
    lipids_g REAL,
    saturated_fat_acids_g REAL,
    monounsaturated_fatty_acids_g REAL,
    polynsaturated_fatty_acids_g REAL,
    linoleic_acid_g REAL,
    trans_fat_acids_g REAL,
    carbohydrates_g REAL,
    sugars_g REAL,
    oligoscarids_g REAL,
    aimido_g REAL,
    fiber_g REAL,
    proteins_g REAL,
    sal___g REAL,
    alcohol_g REAL,
    water_g REAL,
    organic_acids_g REAL,
    colesterol__mg REAL,
    vitamin_a_μg REAL,
    carotene_μg REAL,
    vitamin_d_μg REAL,
    alpha_tocoferol_mg REAL,
    tiamina__mg REAL,
    riboflavina_mg REAL,
    niacina_mg REAL,
    niacin_equivalents_mg REAL,
    triptofano_60__mg REAL,
    vitamin_bs_hopper REAL,
    vitamina_b1_μg REAL,
    vitamina_c_mg REAL,
    folates_μg REAL,
    grey_g REAL,
    sodium_mg REAL,
    potassium_mg REAL,
    calcium_mg REAL,
    phosphor_mg REAL,
    magnesium_mg REAL,
    ferro__mg REAL,
    zinco__mg REAL);""")
    
# commit changes and close database connect
conn.commit()
```

```python
c.execute('''CREATE TABLE IF NOT EXISTS user(
    user_id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    password TEXT,
    e_mail TEXT,
    intolerance_allergy TEXT,
    food_preference TEXT);
    ''')
# commit changes and close database connect
conn.commit()
```

```python
c.execute('''CREATE TABLE IF NOT EXISTS recipes_hist(
    recipes_hist_id INTEGER PRIMARY KEY AUTOINCREMENT,
    user_id INTEGER,
    receipe_id INTEGER,
    rating REAL,
    comments TEXT,
    time_stamp TEXT DEFAULT CURRENT_TIMESTAMP,
    FOREIGN KEY(user_id) REFERENCES user(user_id),
    FOREIGN KEY(receipe_id) REFERENCES recipes(recipe_id));
    ''')
# commit changes and close database connect
conn.commit()
```

```python
c.execute('''CREATE TABLE IF NOT EXISTS nutrition_hist(
    nutrition_hist_id INTEGER PRIMARY KEY AUTOINCREMENT,
    user_id REAL,
    food_id REAL,
    quantity REAL,
    time_stamp TEXT DEFAULT CURRENT_TIMESTAMP,
    portion_kcal REAL,
    FOREIGN KEY(user_id) REFERENCES user(user_id),
    FOREIGN KEY(food_id) REFERENCES nutrition_table(food_id));
    ''')
# commit changes and close database connect
conn.commit()
```

# Populate with new recipes from pickle

```python
table = pd.read_pickle('DataSets\edamam_api_12_mediterranean.p')

# Flatten the column recipe:
def flatten_column(data,column):
    return pd.DataFrame(dict(data[column])).T 

recipes = flatten_column(table,'recipe')
recipes.head(2) 
```

```python
recipes['label']
```

```python
# select the regular size images
def regular_images(row):
    row['images'] = row['images']['REGULAR']['url']

recipes.apply(regular_images, axis=1)
recipes['images']
```

```python
col_to_keep = ['label','images', 'url', 'shareAs','yield', 'dietLabels', 'healthLabels', 'cautions','ingredients', 'calories', 'cuisineType','mealType', 'dishType']
recipes = recipes[col_to_keep]
recipes.head()
```

```python
print(len(recipes.columns))
col_names = recipes.columns
col_names
```

```python
recipes = recipes.applymap(str)
```

```python
recipes.dtypes
```

```python
recipes_info = []

def info_tuple(df_):
    for i in range(len(df_)-1):
        tuples = tuple([df_[col][i] for col in col_names])
        recipes_info.append(tuples)
    return recipes_info

recipes_info = info_tuple(recipes)
```

```python
# Insert many values with parameter queries to prevent sql injections
conn = sqlite3.connect('final_project.db')
c = conn.cursor()
c.executemany("""INSERT INTO recipes (label,images,url,shareAs,yield,dietLabels,healthLabels,cautions,ingredients,calories,cuisineType,mealType,dishType)
                 VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?);""", 
                 recipes_info)
# commit changes and close database connect
conn.commit()
conn.close()
```
