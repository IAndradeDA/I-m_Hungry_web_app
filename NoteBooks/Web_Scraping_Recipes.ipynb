```python
from bs4 import BeautifulSoup
import pandas as pd
import requests  
```

```python
url = 'https://www.eatingwell.com/article/8005469/no-sugar-weight-loss-meal-plan/'

response = requests.get(url)
response.status_code
```

```python
response.content
```

```python
soup = BeautifulSoup(response.content)
soup 
```

```python
soup.find_all('h3')
```

```python
soup.find_all('h3')[14].text
```

```python
meals = [meal.text.strip() for meal in soup.find_all('h3')[14:]]
meals
```

```python
soup.find_all('li')
```

```python
len(soup.find_all('li'))
```

```python
for item in soup.find_all('li')[160:]:
    print(item)
```

```python
recipes_url = []

for item in soup.find_all('li')[160:]:
    try:
        recipe = item.text.strip()
        url = item.find('a').get('href')
        recipes_url.append({'recipe':recipe, 'url':url})
    except:
        recipe = item.text.strip()
        url = 0
        recipes_url.append({'recipe':recipe, 'url':url})
```

```python
recipes_url_pdf = pd.DataFrame(recipes_url)
```
